### Conditionally adding columns

```python
df["miles_category"]=np.where(df["MILES*"]>10,"Long Trip", "Short Trip")
df.drop(columns=["miles_cat"])
```

### Grouping=SAC(Splitting, Applying, Combining)

```python
df.groupby("START*")["MILES*"].agg("mean")
```

```python
df.groupby("START*")["MILES*"].mean()
```

```python
df.groupby("START*").mean()
```

```python
df4=df.groupby(["START*", "STOP*"]).agg(["mean", "sum"]).head(10)
df4.columns
```

```python
df.groupby("START_DATE*").agg("mean")
```

```python
df=df.iloc[:-1,:]
df.tail()
df["START_DATE*"]=pd.to_datetime(df["START_DATE*"])
df.dtypes
```

```python
result=df.groupby(["START*"]).agg({"START_DATE*":["min","max"],"MILES*":"mean"})
result
```

```python
result=df.groupby("START*").agg({"MILES*":"mean","START_DATE*":["min","max"]})
result.reset_index(inplace=True)
result
result.columns=["START_CITY*","AVG_DIST*","EARLIEST_TRIP*","RECENT_TRIP*"]
result
```

```python
result.to_csv("index.csv")
```
